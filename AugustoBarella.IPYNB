Importações

```python
import csv 
import numpy as np
import pandas as pd 
import requests
import seaborn as sns
import matplotlib.pyplot as plt
```

Abrindo o Arquivo com Pandas. 


```python
arquivo = pd.read_csv('D.SDA.PDA.001.CON.202302.csv', encoding='latin1', sep=';', decimal=',')
```

Algumas configurações iniciais(conversões de dados).

```python
arquivo['Dt Nascimento'] = pd.to_datetime(arquivo['Dt Nascimento'], format='%d/%m/%Y')
arquivo['Espécie'] = arquivo['Espécie'].astype('string')
arquivo['CID'] = arquivo['CID'].astype('string')
arquivo['Despacho'] = arquivo['Despacho'].astype('string')
arquivo['Sexo.'] = arquivo['Sexo.'].astype('string')
arquivo['Clientela'] = arquivo['Clientela'].astype('string')
arquivo['Tipo de Cálculo'] = arquivo['Tipo de Cálculo'].astype('string')
arquivo['Mun Resid'] = arquivo['Mun Resid'].astype('string')
arquivo['Vínculo dependentes'] = arquivo['Vínculo dependentes'].astype('string')
arquivo['UF'] = arquivo['UF'].astype('string')
arquivo['Forma Filiação'] = arquivo['Forma Filiação'].astype('string')
#arquivo['Dt Nascimento'].dt.strftime('%d-%m-%Y')
```

```python
#arquivo.info()
#print("descrição do arquivo")
#arquivo.describe()
```

Modelando os dados para fazer a analise 

```python
#Especie

coluna_especie  = arquivo['Espécie']

valores_especie, contagens_especie = np.unique(coluna_especie, return_counts=True)

df_especies = pd.DataFrame({'valores': valores_especie, 'contagens': contagens_especie})

#CID

coluna_CID  = arquivo['CID']

valores_CID, contagens_CID = np.unique(coluna_CID, return_counts=True)

df_CID = pd.DataFrame({'valores': valores_CID, 'contagens': contagens_CID})

#Despacho

coluna_despacho  = arquivo['Despacho']

valores_despacho, contagens_despacho = np.unique(coluna_despacho, return_counts=True)

df_despacho = pd.DataFrame({'valores': valores_despacho, 'contagens': contagens_despacho})

#Sexo

coluna_sexo = arquivo['Sexo.']

valores_sexo, contagens_sexo = np.unique(coluna_sexo, return_counts=True)

df_sexo = pd.DataFrame({'valores': valores_sexo, 'contagens': contagens_sexo})

#Clientela

coluna_clientela  = arquivo['Clientela']

valores_clientela, contagens_clientela = np.unique(coluna_clientela, return_counts=True)

df_clientela = pd.DataFrame({'valores': valores_clientela, 'contagens': contagens_clientela})


#Tipo de Calculo

coluna_calculo  = arquivo['Tipo de Cálculo']

valores_calculo, contagens_calculo = np.unique(coluna_calculo, return_counts=True)

df_calculo = pd.DataFrame({'valores': valores_calculo, 'contagens': contagens_calculo})

#Mun Resid

coluna_resid  = arquivo['Mun Resid']

valores_resid, contagens_resid = np.unique(coluna_resid, return_counts=True)

df_resid = pd.DataFrame({'valores': valores_resid, 'contagens': contagens_resid})

#Vínculos dependentes

coluna_dependentes  = arquivo['Vínculo dependentes']

valores_dependentes, contagens_dependentes = np.unique(coluna_dependentes, return_counts=True)

df_dependentes = pd.DataFrame({'valores': valores_dependentes, 'contagens': contagens_dependentes})

#Forma Filiação

coluna_forma  = arquivo['Forma Filiação']

valores_forma, contagens_forma = np.unique(coluna_forma, return_counts=True)

df_forma = pd.DataFrame({'valores': valores_forma, 'contagens': contagens_forma})

#UF

coluna_UF  = arquivo['UF']

valores_UF, contagens_UF = np.unique(coluna_UF, return_counts=True)

df_UF = pd.DataFrame({'valores': valores_UF, 'contagens': contagens_UF})

#QT SM RMI

coluna_rmi  = arquivo['Qt SM RMI']

valores_rmi, contagens_rmi = np.unique(coluna_rmi, return_counts=True)

df_rmi = pd.DataFrame({'valores': valores_rmi, 'contagens': contagens_rmi})


```

Estatisticas sobre o CID (classificação internacional de doenças)

```python

for valor, contagem in zip(valores_CID, contagens_CID):
    print(f'{valor}: {contagem}')
soma_total = sum(contagens_CID)
print(soma_total)
```

Como podemos ver temos varias doenças, mas podemos observar um fato interessante: dados em branco, zerados e não classificados

```python
em_branco = "Em Branco                                    "
zerados = "Zerados                                      "
nao_classificado = "{ñ class}                                    "

for valor, contagem in zip(valores_CID, contagens_CID):
    if valor == em_branco or valor == zerados or valor == nao_classificado:
        print(f'{valor}: {contagem}')

print("Total de CID:")
print(soma_total)
```

Pode se observar que 4,1975% dos CID são registradas como Em branco.
47,5288% são Zerados.
0,1431% são de "{ñ class}".
Ou seja mais da metade dos CID não são bem especificados.
